```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# cargar el dataset
df = pd.read_csv('data/train.csv')
```

```python
# 1. Distribucion por edades
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Distribución de la Edad de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.savefig('visualizaciones/distribucion_edad.png')
plt.show()  # Display the plot
plt.close()
```

Este gráfico tiene como objetivo visualizar la distribución de las edades de los pasajeros a bordo del Titanic. Se utiliza un histograma que muestra cuántos pasajeros pertenecen a diferentes rangos de edad.
#Interpretación: Al observar el histograma, se puede notar que la mayoría de los pasajeros eran adultos, con una concentración significativa en el rango de 20 a 40 años. Sin embargo, también se observa una cantidad considerable de niños, especialmente menores de 10 años. Esto sugiere que, aunque había muchos adultos, los niños también estaban representados en el viaje.

```python
# 2. Supervivencia por Género
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Supervivencia por Género')
plt.xlabel('Género')
plt.ylabel('Número de Pasajeros')
plt.legend(title='Supervivió', labels=['No', 'Sí'])
plt.savefig('visualizaciones/supervivencia_genero.png')
plt.show()  
plt.close()
```

```python
Este gráfico compara la tasa de supervivencia entre hombres y mujeres, utilizando un gráfico de barras para mostrar cuántos sobrevivieron en cada grupo.
Interpretación: Los resultados indican que las mujeres tenían una tasa de supervivencia significativamente más alta en comparación con los hombres. Esto sugiere que el género fue un factor importante en la supervivencia durante el desastre del Titanic, posiblemente debido a las normas sociales de la época que priorizaban salvar a las mujeres y niños.
```

```python
# 3. Supervivencia por Clase de Pasajero
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Supervivencia por Clase de Pasajero')
plt.xlabel('Clase de Pasajero')
plt.ylabel('Número de Pasajeros')
plt.legend(title='Supervivió', labels=['No', 'Sí'])
plt.savefig('visualizaciones/supervivencia_clase.png')
plt.show()  
plt.close()
```

```python
# 4. Relación entre Edad y Tarifa (Fare)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', alpha=0.6)
plt.title('Relación entre Edad y Tarifa')
plt.xlabel('Edad')
plt.ylabel('Tarifa (Fare)')
plt.savefig('visualizaciones/edad_vs_tarifa.png')
plt.show()  
plt.close()
```

```python
# 5. mapa de Correlaciones
# Seleccionar solo las columnas numéricas
numerical_cols = df.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
correlation_matrix = numerical_cols.corr()

# Crear un mapa de calor de correlaciones
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Mapa de Calor de Correlaciones entre Variables Numéricas')
plt.show()  # Mostrar el gráfico
```
